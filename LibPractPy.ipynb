```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/gheniabla/datasets/master/life_exp_income_population.csv')
type(data)
#display(data)
data.head()
sns.set_style(style= "white")

g = sns.scatterplot(data=data, x="income", y="life_exp",sizes=(30, 220), hue="region", size="population", alpha=0.8)
#s= 200
g.legend(loc='center right', bbox_to_anchor=(1.10, 0.5))


#g.legend(multialignment = "left")
#g.set_bbox_to_anchor((1.05, 1))
sns.despine(left=True, bottom = True)

sns.set(rc={"figure.figsize":(15, 8), 'legend.frameon':False, }) #width=8, height=4

plt.xlabel('Income')
plt.ylabel('Life Expectancy')

plt.grid()

plt.xscale('log')

ax = plt.gca()
ax.set_title("Hunter Snyder's Income vs. Life Expectancy by Region & Population Plot")
#---show the plot---
plt.show()
```
